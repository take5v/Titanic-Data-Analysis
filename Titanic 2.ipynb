```python
#%pylab inline
import pandas as pd
import numpy as np
import scipy as sp
from patsy import dmatrices
# import the machine learning library that holds the randomforest
from sklearn.ensemble import RandomForestRegressor
from sklearn.grid_search import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
t = pd.read_csv('train.csv', header=0)
t = t.drop('Cabin', axis=1)
t = t.dropna()
print t.count()
test_data = pd.read_csv("test.csv")
test_data = test_data.drop('Cabin', axis=1)
test_data[test_data.Age.isnull()] = test_data.Age.mean()
test_data[test_data.Fare.isnull()] = test_data.Fare.mean()
```

```python
test_data['Survived'] = -1
t_full = pd.concat([t, test_data])

formula_ml = 'Survived ~ C(Pclass) + C(Sex) + Age + SibSp + Parch + C(Embarked)'
y, x = dmatrices(formula_ml, data=t_full, return_type='dataframe')

parameter_grid = {
    #"kernel": ["poly"],
    "kernel": ["linear", "rbf"],
    "gamma": [0.0, 0.1, 0.5, 0.7, 0.9, 1.5, 2.4, 3.0]
}
                                            
classifier = svm.SVC()
grid_search = GridSearchCV(classifier, parameter_grid, n_jobs=-1, cv=10)
grid_search.fit(x[y.Survived != -1], np.asarray(y[y.Survived != -1]).ravel())
best_params = sorted(grid_search.grid_scores_, key=lambda x: x[1], reverse=True)[:10]
best_params

```
